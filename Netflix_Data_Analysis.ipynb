```python
tv_show = df[df["type"] == "TV Show"].copy()
tv_show["date_added"] = pd.to_datetime(tv_show["date_added"])
most_recent_tv_show = tv_show.sort_values(by="date_added", ascending=False).head()
print(most_recent_tv_show["title"])
```
