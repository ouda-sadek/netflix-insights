```python
# Convertir la colonne 'duration' en chaîne de caractères
df['duration'] = df['duration'].astype(str)

# Pour les films : supprimer " min" et convertir en entier
df['duration'] = df['duration'].str.replace(' min', '', regex=False)

# Pour les séries : extraire le nombre de saisons, par exemple "2 Seasons" devient 2
df['duration'] = df['duration'].str.replace(' Seasons', '', regex=False)

# Convertir la colonne 'duration' en entier
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')

# Vérifier les 5 premières lignes pour s'assurer que la conversion a bien fonctionné
print(df['duration'].head())

```

```python
print(df[df['duration'].isnull()])
```

```python
# Modifier la variable 'listed_in' pour la transformer en liste de chaînes de caractères
df['listed_in'] = df['listed_in'].str.split(', ')

# Vérifier les 5 premières lignes pour s'assurer que la conversion a bien fonctionné
print(df['listed_in'].head())
```

```python
# Afficher les valeurs uniques de chaque variable
print("Valeurs uniques de 'type':")
print(df['type'].unique())
print("\n")

print("Valeurs uniques de 'country':")
print(df['country'].unique())
print("\n")

print("Valeurs uniques de 'release_year':")
print(df['release_year'].unique())
print("\n")

print("Valeurs uniques de 'rating':")
print(df['rating'].unique())
print("\n")

# "Exploser" la colonne 'listed_in' pour chaque genre dans les listes
exploded_listed_in = df['listed_in'].explode()

# Afficher les valeurs uniques de 'listed_in'
print("Valeurs uniques de 'listed_in':")
print(exploded_listed_in.unique())

```

```python
# Grouper par 'director' et compter le nombre d'œuvres produites par chaque réalisateur
director_counts = df.groupby('director')['title'].count()

# Afficher les réalisateurs ayant produit plus d'une œuvre
directors_multiple_works = director_counts[director_counts > 1]

# Afficher les réalisateurs et le nombre d'œuvres qu'ils ont produites
print(directors_multiple_works)
```

```python
# Extraire l'année de la colonne 'date_added'
df['year_added'] = df['date_added'].dt.year

# Compter le nombre de films ajoutés par année
added_per_year = df['year_added'].value_counts()

# Afficher l'année avec le plus grand nombre d'ajouts
most_added_year = added_per_year.idxmax()
most_added_count = added_per_year.max()

print(f"L'année avec le plus de films ajoutés est {most_added_year}, avec {most_added_count} films.")

```

```python
# Extraire l'année de la colonne 'date_added' si ce n'est pas déjà fait
df['year_added'] = df['date_added'].dt.year

# Filtrer uniquement les séries
df_series = df[df['type'] == 'TV Show']

# Compter le nombre de séries ajoutées par année
series_per_year = df_series['year_added'].value_counts()

# Trouver l'année avec le plus grand nombre de séries ajoutées
most_added_year = series_per_year.idxmax()
most_added_count = series_per_year.max()

print(f"L'année avec le plus de séries ajoutées est {most_added_year} avec {most_added_count} séries.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisation de la répartition du type d'œuvres (Movie vs TV Show)
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=df)
plt.title('Répartition du type d\'œuvres')
plt.xlabel('Type d\'œuvre')
plt.ylabel('Nombre')
plt.show()
# Créer une nouvelle série avec les pays de toutes les œuvres
countries = df['country'].dropna().str.split(', ').explode()

# Visualisation de la répartition des pays
plt.figure(figsize=(12, 10))  # Augmenter la taille du graphique
sns.countplot(y=countries, order=countries.value_counts().index)
plt.title('Répartition des œuvres en fonction des pays')
plt.xlabel('Nombre d\'œuvres')
plt.ylabel('Pays')

# Faire une rotation des étiquettes pour mieux les voir
plt.xticks(rotation=90)
plt.show()
```

```python
# Créer une nouvelle série avec les pays de toutes les œuvres
countries = df['country'].dropna().str.split(', ').explode()

# Compter les occurrences des pays et sélectionner les 15 pays les plus fréquents
top_countries = countries.value_counts().head(15)

# Visualisation des pays les plus fréquents
plt.figure(figsize=(12, 8))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 15 des pays avec le plus d\'œuvres')
plt.xlabel('Nombre d\'œuvres')
plt.ylabel('Pays')
plt.show()

```

```python
# Visualisation de la répartition des années
plt.figure(figsize=(12, 8))
sns.countplot(x='year_added', data=df)
plt.title('Répartition des œuvres en fonction des années')
plt.xlabel('Année d\'ajout')
plt.ylabel('Nombre d\'œuvres')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualisation de la répartition des ratings
plt.figure(figsize=(12, 8))
sns.countplot(x='rating', data=df, order=df['rating'].value_counts().index)
plt.title('Répartition des évaluations (ratings) des œuvres')
plt.xlabel('Évaluation')
plt.ylabel('Nombre d\'œuvres')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filtrer les séries uniquement
series = df[df['type'] == 'TV Show']

# Visualisation de la répartition de la durée des séries
plt.figure(figsize=(12, 8))
sns.histplot(series['duration'], bins=10, kde=True)
plt.title('Répartition de la durée des séries (en saisons)')
plt.xlabel('Nombre de saisons')
plt.ylabel('Nombre de séries')
plt.show()
```

```python
# "Exploser" la colonne 'listed_in' pour chaque genre dans les listes
genres = df['listed_in'].explode()

# Visualisation de la répartition des genres
plt.figure(figsize=(12, 8))
sns.countplot(y=genres, order=genres.value_counts().index)
plt.title('Répartition des genres d\'œuvres')
plt.xlabel('Nombre d\'œuvres')
plt.ylabel('Genre')
plt.show()
```

```python
# Filtrer les séries uniquement
series = df[df['type'] == 'TV Show']

# Trier les séries par durée (nombre de saisons) de manière décroissante
top_5_series = series.sort_values(by='duration', ascending=False).head(5)

# Visualisation du top 5 des séries les plus longues
plt.figure(figsize=(12, 8))
sns.barplot(x='duration', y='title', data=top_5_series, palette='viridis')
plt.title('Top 5 des séries les plus longues')
plt.xlabel('Nombre de saisons')
plt.ylabel('Titre de la série')
plt.show()
```

```python
# Filtrer les films uniquement
movies = df[df['type'] == 'Movie']

# Trier les films par durée (en minutes) de manière décroissante
top_5_movies = movies.sort_values(by='duration', ascending=False).head(5)

# Visualisation du top 5 des films les plus longs
plt.figure(figsize=(12, 8))
sns.barplot(x='duration', y='title', data=top_5_movies, palette='viridis')
plt.title('Top 5 des films les plus longs')
plt.xlabel('Durée (en minutes)')
plt.ylabel('Titre du film')
plt.show()
```
