```python
films_df = dg[dg['type'] == 'Movie'].copy()


films_df["date_added"] = pd.to_datetime(films_df["date_added"]).dt.year

movie_per_year = films_df["date_added"].value_counts()

year_movie_max = movie_per_year.idxmax()
max_movie = movie_per_year.max()

print(f"the year with the most added movie is {year_movie_max} with {max_movie} movies.")
```

```python
###!pip install matplotlib
import matplotlib.pyplot as plt
```

```python
repartition = df["type"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(repartition, labels=repartition.index,autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff'])
plt.title("percentage of movie and tv show")
plt.show()
```

```python
dc = df["country"].copy()
dc_split = dc.str.split(",")
explode_country = dc_split.explode()
repartition = explode_country.value_counts()

seuil = 100
repartition_optimise = repartition[repartition >= seuil]
autres = repartition[repartition < seuil].sum()
repartition_optimise["Autres"] = autres

plt.bar(repartition_optimise.index, repartition_optimise.values)


plt.xlabel("country")
plt.ylabel("frequencies")
plt.title("Distribution of works by country (with 'Others' for less frequent countries)")
plt.xticks(rotation=90)  
plt.show()
```

```python
year_df = dg.copy()
year_df["date_added"] = pd.to_datetime(year_df["date_added"]).dt.year
year_repartition = year_df["date_added"].value_counts()


plt.bar(year_repartition.index, year_repartition.values)
plt.xlabel("year")
plt.ylabel("number of work")
plt.title("number of works per year")
plt.xticks(rotation=90)  
plt.show()
```
